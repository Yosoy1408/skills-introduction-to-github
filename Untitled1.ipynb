```python
def start(update: Update, context: CallbackContext) -> None:
    user = update.effective_user
    mensaje = game.crear_usuario(user.id, user.first_name)

    keyboard = [
        [InlineKeyboardButton("üèóÔ∏è Construir", callback_data='construir')],
        [InlineKeyboardButton("üë§ Mi Perfil", callback_data='perfil')],
        [InlineKeyboardButton("üèÜ Logros", callback_data='logros')]
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)

    update.message.reply_text(
        f"{mensaje}\n\n"
        "¬øQu√© te gustar√≠a hacer?",
        reply_markup=reply_markup
    )

def button(update: Update, context: CallbackContext) -> None:
    query = update.callback_query
    query.answer()

    if query.data == 'construir':
        respuesta = game.construir(query.from_user.id)
        query.edit_message_text(text=respuesta)
    elif query.data == 'perfil':
        respuesta = game.ver_perfil(query.from_user.id)
        query.edit_message_text(text=respuesta, parse_mode=ParseMode.MARKDOWN)
    elif query.data == 'logros':
        usuario = game.usuarios.get(query.from_user.id)
        if usuario:
            respuesta = "üèÜ *Tus logros:*\n\n"
            if usuario['logros']:
                for logro in usuario['logros']:
                    logro_info = game.logros[logro]
                    respuesta += f"‚Ä¢ {logro_info['nombre']}: {logro_info['descripcion']}\n"
            else:
                respuesta += "A√∫n no has conseguido ning√∫n logro. ¬°Sigue jugando!"
        else:
            respuesta = "‚ùå Necesitas registrarte primero con /start"
        query.edit_message_text(text=respuesta, parse_mode=ParseMode.MARKDOWN)
```

```python
def main() -> None:
    try:
        updater = Updater(TOKEN)
        dp = updater.dispatcher

        dp.add_handler(CommandHandler("start", start))
        dp.add_handler(CallbackQueryHandler(button))

        updater.start_polling()
        print("¬°Bot iniciado correctamente! üöÄ")
        print("Usa Ctrl+C para detener el bot")
        updater.idle()
    except Exception as e:
        print(f"Error al iniciar el bot: {e}")

if __name__ == '__main__':
    main()
```
