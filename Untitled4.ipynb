```python
def setup_database():
    conn = sqlite3.connect('dreambuilder.db')
    c = conn.cursor()

    # Tabla de usuarios
    c.execute('''
        CREATE TABLE IF NOT EXISTS usuarios (
            user_id INTEGER PRIMARY KEY,
            username TEXT,
            nivel INTEGER DEFAULT 1,
            experiencia INTEGER DEFAULT 0,
            monedas INTEGER DEFAULT 1000,
            casas_construidas INTEGER DEFAULT 0,
            fecha_registro DATETIME DEFAULT CURRENT_TIMESTAMP,
            ultimo_bonus DATETIME,
            ultimo_evento DATETIME,
            evento_activo TEXT,
            referido_por INTEGER,
            total_donaciones REAL DEFAULT 0
        )
    ''')

    # Tabla de logros
    c.execute('''
        CREATE TABLE IF NOT EXISTS logros_usuarios (
            user_id INTEGER,
            logro_id TEXT,
            fecha_logro DATETIME DEFAULT CURRENT_TIMESTAMP,
            PRIMARY KEY (user_id, logro_id)
        )
    ''')

    # Tabla de construcciones
    c.execute('''
        CREATE TABLE IF NOT EXISTS construcciones (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER,
            nivel_construccion INTEGER,
            fecha_inicio DATETIME,
            fecha_fin DATETIME,
            estado TEXT,
            recompensa_reclamada BOOLEAN DEFAULT FALSE
        )
    ''')

    # Tabla de donaciones
    c.execute('''
        CREATE TABLE IF NOT EXISTS donaciones (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER,
            cantidad REAL,
            fecha DATETIME DEFAULT CURRENT_TIMESTAMP,
            metodo_pago TEXT,
            estado TEXT
        )
    ''')

    # Tabla de eventos
    c.execute('''
        CREATE TABLE IF NOT EXISTS eventos (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            tipo_evento TEXT,
            fecha_inicio DATETIME,
            fecha_fin DATETIME,
            estado TEXT
        )
    ''')

    # Tabla de mensajes diarios
    c.execute('''
        CREATE TABLE IF NOT EXISTS mensajes_diarios (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id INTEGER,
            mensaje TEXT,
            fecha DATETIME DEFAULT CURRENT_TIMESTAMP
        )
    ''')

    conn.commit()
    conn.close()
    print("✅ Base de datos configurada")

setup_database()
```

```python
class DreamBuilderGame:
    def __init__(self):
        self.construcciones_activas = {}
        self.eventos_activos = {}

    def iniciar_usuario(self, user_id, username, referido_por=None):
        conn = sqlite3.connect('dreambuilder.db')
        c = conn.cursor()

        try:
            c.execute("SELECT * FROM usuarios WHERE user_id = ?", (user_id,))
            if not c.fetchone():
                c.execute(
                    "INSERT INTO usuarios (user_id, username, referido_por) VALUES (?, ?, ?)",
                    (user_id, username, referido_por)
                )
                conn.commit()
                print(f"✅ Nuevo usuario registrado: {username}")
        except Exception as e:
            print(f"❌ Error al registrar usuario: {e}")
        finally:
            conn.close()

    def obtener_estado_usuario(self, user_id):
        conn = sqlite3.connect('dreambuilder.db')
        c = conn.cursor()

        c.execute("SELECT * FROM usuarios WHERE user_id = ?", (user_id,))
        usuario = c.fetchone()
        conn.close()

        if usuario:
            return {
                "user_id": usuario[0],
                "username": usuario[1],
                "nivel": usuario[2],
                "experiencia": usuario[3],
                "monedas": usuario[4],
                "casas_construidas": usuario[5],
                "fecha_registro": usuario[6],
                "ultimo_bonus": usuario[7],
                "ultimo_evento": usuario[8],
                "evento_activo": usuario[9],
                "referido_por": usuario[10],
                "total_donaciones": usuario[11]
            }
        return None

    def actualizar_progreso(self, user_id, monedas=0, experiencia=0, casas=0):
        conn = sqlite3.connect('dreambuilder.db')
        c = conn.cursor()

        try:
            c.execute(
                "UPDATE usuarios SET monedas = monedas + ?, experiencia = experiencia + ?, casas_construidas = casas_construidas + ? WHERE user_id = ?",
                (monedas, experiencia, casas, user_id)
            )
            conn.commit()
        except Exception as e:
            print(f"❌ Error al actualizar progreso: {e}")
        finally:
            conn.close()

    def verificar_logros(self, user_id):
        usuario = self.obtener_estado_usuario(user_id)
        if not usuario:
            return None

        logros_obtenidos = []
        if usuario["casas_construidas"] == 1:
            logros_obtenidos.append(self.otorgar_logro(user_id, "constructor_novato"))
        elif usuario["casas_construidas"] == 5:
            logros_obtenidos.append(self.otorgar_logro(user_id, "constructor_experto"))
        elif usuario["casas_construidas"] == 10:
            logros_obtenidos.append(self.otorgar_logro(user_id, "constructor_maestro"))

        return logros_obtenidos

    def otorgar_logro(self, user_id, logro_id):
        conn = sqlite3.connect('dreambuilder.db')
        c = conn.cursor()

        try:
            c.execute(
                "SELECT * FROM logros_usuarios WHERE user_id = ? AND logro_id = ?",
                (user_id, logro_id)
            )
            if not c.fetchone():
                c.execute(
                    "INSERT INTO logros_usuarios (user_id, logro_id) VALUES (?, ?)",
                    (user_id, logro_id)
                )
                conn.commit()
                return LOGROS[logro_id]["nombre"]
        except Exception as e:
            print(f"❌ Error al otorgar logro: {e}")
        finally:
            conn.close()
        return None

    def activar_evento(self, tipo_evento):
        if tipo_evento not in EVENTOS:
            return False

        evento = EVENTOS[tipo_evento]
        fecha_inicio = datetime.now()
        fecha_fin = fecha_inicio + timedelta(seconds=evento["duracion"])

        conn = sqlite3.connect('dreambuilder.db')
        c = conn.cursor()

        try:
            c.execute(
                "INSERT INTO eventos (tipo_evento, fecha_inicio, fecha_fin, estado) VALUES (?, ?, ?, ?)",
                (tipo_evento, fecha_inicio, fecha_fin, "activo")
            )
            conn.commit()
            self.eventos_activos[tipo_evento] = {
                "fecha_inicio": fecha_inicio,
                "fecha_fin": fecha_fin
            }
            return True
        except Exception as e:
            print(f"❌ Error al activar evento: {e}")
        finally:
            conn.close()

    def verificar_eventos_activos(self):
        conn = sqlite3.connect('dreambuilder.db')
        c = conn.cursor()

        c.execute("SELECT * FROM eventos WHERE estado = 'activo'")
        eventos = c.fetchall()
        conn.close()

        eventos_activos = []
        for evento in eventos:
            tipo_evento = evento[1]
            fecha_fin = datetime.strptime(evento[3], "%Y-%m-%d %H:%M:%S")
            if datetime.now() > fecha_fin:
                self.finalizar_evento(evento[0])
            else:
                eventos_activos.append(tipo_evento)

        return eventos_activos

    def finalizar_evento(self, evento_id):
        conn = sqlite3.connect('dreambuilder.db')
        c = conn.cursor()

        try:
            c.execute(
                "UPDATE eventos SET estado = 'finalizado' WHERE id = ?",
                (evento_id,)
            )
            conn.commit()
        except Exception as e:
            print(f"❌ Error al finalizar evento: {e}")
        finally:
            conn.close()
```

```python
def registrar_donacion(user_id, cantidad, metodo_pago="Desconocido"):
    conn = sqlite3.connect('dreambuilder.db')
    c = conn.cursor()

    try:
        c.execute(
            "INSERT INTO donaciones (user_id, cantidad, metodo_pago, estado) VALUES (?, ?, ?, ?)",
            (user_id, cantidad, metodo_pago, "completado")
        )
        c.execute(
            "UPDATE usuarios SET total_donaciones = total_donaciones + ? WHERE user_id = ?",
            (cantidad, user_id)
        )
        conn.commit()
        return True
    except Exception as e:
        print(f"❌ Error al registrar donación: {e}")
        return False
    finally:
        conn.close()
```

```python
# Enviar notificaciones automáticas a los usuarios
def enviar_notificaciones(context: CallbackContext):
    conn = sqlite3.connect('dreambuilder.db')
    c = conn.cursor()

    c.execute("SELECT user_id FROM usuarios")
    usuarios = c.fetchall()
    conn.close()

    for usuario in usuarios:
        user_id = usuario[0]
        context.bot.send_message(
            chat_id=user_id,
            text="📢 *Notificación:* ¡No olvides construir hoy y ayudar a más familias! 🏠",
            parse_mode=ParseMode.MARKDOWN
        )
```

```python
# Comando /start
def start(update: Update, context: CallbackContext):
    user = update.effective_user
    game.iniciar_usuario(user.id, user.username)

    keyboard = [
        [InlineKeyboardButton("🏠 Construir Casa", callback_data='construir')],
        [InlineKeyboardButton("👤 Mi Perfil", callback_data='perfil')],
        [InlineKeyboardButton("🎮 Mini-Juegos", callback_data='minijuegos')],
        [InlineKeyboardButton("🏆 Logros", callback_data='logros')],
        [InlineKeyboardButton("📊 Estadísticas", callback_data='estadisticas')],
        [InlineKeyboardButton("❓ Ayuda", callback_data='ayuda')]
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)

    welcome_message = (
        f"¡Hola {user.first_name}! 👋\n\n"
        "Bienvenido a *DreamBuilder: Construyendo Esperanzas* 🏠\n\n"
        "Este bot te permite:\n"
        "• Construir casas virtuales 🏗\n"
        "• Ganar recompensas 🎁\n"
        "• Ayudar a familias reales 👨‍👩‍👧‍👦\n"
        "• Participar en eventos especiales 🎉\n\n"
        "¿Qué te gustaría hacer?"
    )

    update.message.reply_text(
        welcome_message,
        reply_markup=reply_markup,
        parse_mode=ParseMode.MARKDOWN
    )

# Comando /help
def help_command(update: Update, context: CallbackContext):
    help_text = (
        "*Comandos disponibles:*\n\n"
        "/start - Iniciar el bot\n"
        "/help - Mostrar este mensaje de ayuda\n"
        "/perfil - Ver tu perfil\n"
        "/estadisticas - Ver estadísticas globales\n"
        "/logros - Ver tus logros\n"
        "/minijuegos - Acceder a los mini-juegos\n"
        "/adivina - Jugar a adivinar el número\n\n"
        "*¿Necesitas más ayuda?*\n"
        "Contacta con el creador: inverydesar@gmail.com"
    )

    update.message.reply_text(
        help_text,
        parse_mode=ParseMode.MARKDOWN
    )

# Comando /perfil
def perfil(update: Update, context: CallbackContext):
    user = update.effective_user
    estado = game.obtener_estado_usuario(user.id)

    if not estado:
        update.message.reply_text(
            "❌ No se encontró tu perfil. Usa /start para crear uno nuevo."
        )
        return

    perfil_text = (
        f"👤 *Perfil de {user.first_name}*\n\n"
        f"🏅 Nivel: {estado['nivel']}\n"
        f"⭐️ Experiencia: {estado['experiencia']}\n"
        f"💰 Monedas: {estado['monedas']}\n"
        f"🏠 Casas construidas: {estado['casas_construidas']}\n"
        f"💎 Total donaciones: ${estado['total_donaciones']}\n"
        f"📅 Fecha de registro: {estado['fecha_registro']}\n\n"
        "Usa los botones para más acciones:"
    )

    keyboard = [
        [InlineKeyboardButton("🏠 Construir Casa", callback_data='construir')],
        [InlineKeyboardButton("🎮 Mini-Juegos", callback_data='minijuegos')],
        [InlineKeyboardButton("🏆 Ver Logros", callback_data='logros')],
        [InlineKeyboardButton("⬅️ Volver", callback_data='volver')]
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)

    update.message.reply_text(
        perfil_text,
        reply_markup=reply_markup,
        parse_mode=ParseMode.MARKDOWN
    )
```

```python
def main():
    global game
    game = DreamBuilderGame()

    updater = Updater(TOKEN)
    dp = updater.dispatcher

    # Comandos básicos
    dp.add_handler(CommandHandler("start", start))
    dp.add_handler(CommandHandler("help", help_command))
    dp.add_handler(CommandHandler("perfil", perfil))
    dp.add_handler(CommandHandler("estadisticas", estadisticas))
    dp.add_handler(CommandHandler("logros", logros))
    dp.add_handler(CommandHandler("minijuegos", minijuegos))
    dp.add_handler(CommandHandler("adivina", adivina_numero))

    # Manejo de mensajes para mini-juegos
    dp.add_handler(MessageHandler(Filters.text & ~Filters.command, manejar_respuesta))

    # Manejador de botones
    dp.add_handler(CallbackQueryHandler(button_callback))

    # Tareas automáticas (notificaciones)
    job_queue = updater.job_queue
    job_queue.run_repeating(enviar_notificaciones, interval=86400, first=0)  # Cada 24 horas

    # Iniciar el bot
    updater.start_polling()
    print("✅ Bot iniciado correctamente 🚀")

    updater.idle()

if __name__ == '__main__':
    main()
```
